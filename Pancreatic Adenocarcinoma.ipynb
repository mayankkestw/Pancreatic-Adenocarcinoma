# Pancreatic Cancer

Pancreatic cancer is a type of cancer that starts in the pancreas. (Cancer starts when cells in the body begin to grow out of control. To learn more about how cancers start and spread, see What Is Cancer?)

Pancreatic adenocarcinoma is the most common type of pancreatic cancer. Pancreatic neuroendocrine tumors (NETs) are a less common type and are discussed in Pancreatic Neuroendocrine Tumors.

The pancreas is an organ that sits behind the stomach. It's shaped a bit like a fish with a wide head, a tapering body, and a narrow, pointed tail. In adults it's about 6 inches (15 centimeters) long but less than 2 inches (5 centimeters) wide.

The many types of pancreatic cancer can be divided into two general groups. The vast majority of cases (about 95%) occur in the part of the pancreas which produces digestive enzymes, known as the exocrine component. There are several sub-types of exocrine pancreatic cancers, but their diagnosis and treatment have much in common.

# Exocrine Tumors

Most of the pancreas is made of exocrine cells, and most tumors affecting these cells are called adenocarcinomas. Found in many other cancers, including breast, prostate and lung, adenocarcinomas most often form in glands that secrete fluids. Pancreatic adenocarcinomas most often form in the exocrine cells found in pancreas ducts. Treatment for adenocarcinomas of the pancreas is based on the stage and size of the tumor.

# Neuroendocrine Tumors

These tumors are less common and are most often benign. Though rare, cancer stemming from a pancreatic endocrine tumor (PET) affects the hormone-producing cells. These tumors are also called islet cell tumors or neuroendocrine tumors.

# Step 3 : Principal Component Analysis

Principal component analysis (PCA) is a procedure for reducing the dimensionality of the variable space by representing it with a few orthogonal (uncorrelated) variables that capture most of its variability.
For understanding the concept, I recommend to watch this video : https://www.youtube.com/watch?v=_UVHneBUBW0

Our data contains 183 samples we will need to have 183 different axes to represent the data which is not possible right now, so PCA comes into picture this technique will help us to project the higher-dimensional data onto lower dimension.

```python
# 2D plotting

from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_train = pca.fit_transform(X)
```

```python
X_train.shape
```

As we can see the dimensions are reduced.

```python
plt.figure(figsize =(8, 6))
  
plt.scatter(X_train[:, 0], X_train[:, 1], 0.3, cmap ='plasma') 
  
# labeling x and y axes 
plt.title("Visualisation PCA")
plt.xlabel('First Principal Component') 
plt.ylabel('Second Principal Component') 
```

```python
variance = pca.explained_variance_ratio_

var=np.cumsum(np.round(pca.explained_variance_ratio_, decimals=3)*100)
var
```

The above code has explained_variance_ratio_ it tells us the variance explained by each of the dimension. The first component can represent 79.2% of the variation of our data and the oter component can represent 80.8%.

# Step 4 : Clustering

<h3>K-Means</h3>

```python
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
    kmeans.fit(X_train)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

Looking at the Elbow method plot the relevant choice for k would be either 2 or 3. So we will be plotting both.

```python
kmeans = KMeans(n_clusters=2,init='k-means++',max_iter=300,n_init=10,random_state=0)
ymeans = kmeans.fit_predict(X_train)
```

```python
ymeans
```

```python
plt.scatter(X_train[ymeans==0,0],X_train[ymeans==0,1],s=100,c='red',label='Cluster 1')
plt.scatter(X_train[ymeans==1,0],X_train[ymeans==1,1],s=100,c='blue',label='Cluster 2')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='cyan',label='Centroids')
plt.title('Clusters')
plt.legend()
plt.show()
```

We can see the clusters are somewhat separable.

```python
kmeans = KMeans(n_clusters=3,init='k-means++',max_iter=300,n_init=10,random_state=0)
ymeans = kmeans.fit_predict(X_train)
```

```python
plt.scatter(X_train[ymeans==0,0],X_train[ymeans==0,1],s=100,c='red',label='Cluster 1')
plt.scatter(X_train[ymeans==1,0],X_train[ymeans==1,1],s=100,c='blue',label='Cluster 2')
plt.scatter(X_train[ymeans==2,0],X_train[ymeans==2,1],s=100,c='green',label='Cluster 3')

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='cyan',label='Centroids')
plt.title('Clusters')
plt.legend()
plt.show()
```
